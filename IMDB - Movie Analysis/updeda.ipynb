```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('latestimdb_700_movies_cleaned_final.csv')
df.head()
```

```python
pd.set_option('display.max_rows', 1000)
df
```

```python
language_group = df.groupby('Primary_Language').agg({
    'Rating': 'mean',
    'Votes': 'mean',
    'Gross Worldwide': 'mean'
}).reset_index()
```

```python
#Group by Language (Regional Grouping)
```

```python
plt.figure(figsize=(8,5))
sns.barplot(data=language_group, x='Primary_Language', y='Rating', palette='coolwarm')
plt.title('Average Rating by Primary Language')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(data=language_group, x='Primary_Language', y='Votes', palette='viridis')
plt.title('Average Votes by Primary Language')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(data=language_group, x='Primary_Language', y='Gross Worldwide', palette='')
plt.title('Average Gross Worldwide by Primary Language')
plt.xticks(rotation=45)
plt.show()
```

```python
language_group['Rating'] = language_group['Rating'].round(2)
language_group['Votes'] = language_group['Votes'].round(0).astype(int)
language_group['Gross Worldwide'] = language_group['Gross Worldwide'].round(0).astype(int)

print(language_group)
```

```python
plt.figure(figsize=(7,4))
sns.histplot(df['Rating'], kde=True, bins=20, color='blue')
plt.title('Distribution of Ratings')
plt.show()
```

```python
plt.figure(figsize=(7,4))
sns.histplot(df['Votes'], kde=True, bins=20, color='green')
plt.title('Distribution of Votes')
plt.show()
```

```python
plt.figure(figsize=(7,4))
sns.histplot(df['Gross Worldwide'], kde=True, bins=20, color='red')
plt.title('Distribution of Gross Worldwide')
plt.show()
```

```python
#bivariate
```

```python
plt.figure(figsize=(7,5))
sns.scatterplot(data=df, x='Votes', y='Rating', hue='Primary_Language', palette='tab10')
plt.title('Votes vs Rating by Language')
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.scatterplot(data=df, x='Votes', y='Gross Worldwide', hue='Primary_Language', palette='Set2')
plt.title('Gross Worldwide vs Votes by Language')
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.scatterplot(data=df, x='total_minute', y='Rating', hue='Primary_Language', palette='cool')
plt.title('Duration vs Rating by Language')
plt.show()
```
