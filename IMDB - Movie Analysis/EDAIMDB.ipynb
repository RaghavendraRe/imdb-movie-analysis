```python
# Group by Primary Language and calculate mean for Rating, Votes, and Gross Worldwide
language_group = dff.groupby('Primary_Language').agg({
    'Rating': 'mean',
    'Votes': 'mean',
    'Gross Worldwide': 'mean'
}).reset_index()

# Round the values for better readability
language_group = language_group.round(2)

# Display the summary
print(language_group)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
sns.barplot(data=language_group, x='Primary_Language', y='Rating', palette='coolwarm')
plt.title('Average Rating by Primary Language')
plt.xlabel('Language')
plt.ylabel('Average Rating')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(data=language_group, x='Primary_Language', y='Votes', palette='viridis')
plt.title('Average Votes by Primary Language')
plt.xlabel('Language')
plt.ylabel('Average Votes')
plt.xticks(rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Custom color mapping (assign each language a unique color manually)
custom_colors = {
    'Hindi': '#0d0887',
    'Telugu': '#f0f921',
    'Tamil': '#fca636',
    'Malayalam': '#9c179e',
    'Kannada': '#7201a8',
    'Punjabi': '#bd3786',
    'Bengali': '#3b528b',
    'Marathi': '#5ec962'
}

# Plot
plt.figure(figsize=(10, 6))
sns.barplot(
    data=language_group,
    x='Primary_Language',
    y='Gross Worldwide',
    palette=custom_colors
)

plt.title('Average Gross Worldwide by Primary Language')
plt.xlabel('Language')
plt.ylabel('Average Gross (USD)')
plt.xticks(rotation=45)

# Add value labels on top of each bar
for index, row in language_group.iterrows():
    plt.text(index, row['Gross Worldwide'] + 1e5, f"${int(row['Gross Worldwide']):,}", 
             ha='center', va='bottom', fontsize=9)

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# VIBGYOR + 1 color
vibgyor_colors = {
    'Hindi': '#8F00FF',      # Violet
    'Telugu': '#4B0082',     # Indigo
    'Tamil': '#0000FF',      # Blue
    'Malayalam': '#00FF00',  # Green
    'Kannada': '#FFFF00',    # Yellow
    'Punjabi': '#FFA500',    # Orange
    'Bengali': '#FF0000',    # Red
    'Marathi': '#808080'     # Extra (Gray)
}

plt.figure(figsize=(10, 6))
sns.barplot(
    data=language_group,
    x='Primary_Language',
    y='Gross Worldwide',
    palette=vibgyor_colors
)

plt.title('Average Gross Worldwide by Primary Language')
plt.xlabel('Language')
plt.ylabel('Average Gross (USD)')
plt.xticks(rotation=45)

# Add value labels on bars
for index, row in language_group.iterrows():
    plt.text(index, row['Gross Worldwide'] + 1e5, f"${int(row['Gross Worldwide']):,}", 
             ha='center', va='bottom', fontsize=9)

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Filter out Bengali, Marathi, and Punjabi
filtered_data = language_group[~language_group['Primary_Language'].isin(['Bengali', 'Marathi', 'Punjabi'])]

# Set VIBGYOR color palette for remaining languages
vibgyor_colors = {
    'Hindi': '#8F00FF',      # Violet
    'Telugu': '#4B0082',     # Indigo
    'Tamil': '#0000FF',      # Blue
    'Malayalam': '#00FF00',  # Green
    'Kannada': '#FFFF00'     # Yellow
}

plt.figure(figsize=(9, 6))
barplot = sns.barplot(
    data=filtered_data,
    x='Primary_Language',
    y='Gross Worldwide',
    palette=vibgyor_colors
)

plt.title('Average Gross Worldwide by Primary Language)')
plt.xlabel('Language')
plt.ylabel('Average Gross (USD)')
plt.xticks(rotation=45)

# Add value labels on top of bars (only for shown bars)
for p in barplot.patches:
    height = p.get_height()
    plt.text(p.get_x() + p.get_width() / 2, height + 0.05 * height,
             f"${int(height):,}", ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()
```
