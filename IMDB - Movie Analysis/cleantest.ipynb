```python
def convert_votes(vote_str):
    try:
        vote_str = str(vote_str).strip("()")  
        if 'M' in vote_str:
            return int(float(vote_str.replace('M', '')) * 1000000)
        elif 'K' in vote_str:
            return int(float(vote_str.replace('K', '')) * 1000)
        else:
            return int(vote_str)
    except:
        return np.nan

d1['Votes'] = d1['Votes'].apply(convert_votes)
```

```python
#Extract hours and minutes separately
d1["hours"] = d1["Duration"].str.extract(r'(\d+)h').astype(float)
d1["minutes"] = d1["Duration"].str.extract(r'(\d+)m').astype(float)

# Fill missing values with 0
d1['hours'] = d1['hours'].fillna(0)
d1['minutes'] = d1['minutes'].fillna(0)

#  Calculate total minutes
d1['total_minute'] = d1['hours'] * 60 + d1['minutes']

#  Convert total_minute to integer
d1['total_minute'] = d1['total_minute'].astype(int)

# Drop the helper columns
d1.drop(columns=['hours', 'minutes'], axis=1, inplace=True)

#  Drop the original Duration column
d1.drop(columns=["Duration"], axis=1, inplace=True)
```

```python
#  Fill missing Genres with 'Unknown'
d1['Genres'] = d1['Genres'].fillna('Unknown')

#  Create a function to get last 3 genres
def get_last_genres(genre_str):
    genres_list = genre_str.split(',')
    genres_list = [g.strip() for g in genres_list]  # remove spaces
    if len(genres_list) >= 3:
        return ', '.join(genres_list[-3:])  # take LAST 3
    else:
        return ', '.join(genres_list)  # take all available

#  Apply the function
d1['Main_Genre'] = d1['Genres'].apply(get_last_genres)
```

```python
genre_mapping = {
    'Action Epic': 'Action',
    'One-Person Army Action': 'Action',
    'Cop Drama': 'Crime',
    'Legal Drama': 'Drama',
    'Psychological Drama': 'Drama',
    'Political Drama': 'Drama',
    'Docudrama': 'Drama',
    'Coming-of-Age': 'Drama',
    'Martial Arts': 'Action',
    'Romantic Comedy': 'Comedy',
    'Quirky Comedy': 'Comedy',
    'Buddy Comedy': 'Comedy',
    'Dark Comedy': 'Comedy',
    'Slapstick': 'Comedy',
    'Showbiz Drama': 'Drama',
    'Folk Horror': 'Horror',
    'Supernatural Horror': 'Horror',
    'Psychological Thriller': 'Thriller',
    'Conspiracy Thriller': 'Thriller',
    'Political Thriller': 'Thriller',
    'Suspense Mystery': 'Mystery',
    'Quest': 'Adventure',
    'Adventure Epic': 'Adventure',
    'Historical Epic': 'History',
    'War Epic': 'War',
    'Medical Drama': 'Drama',
    'Crime Documentary': 'Crime',
    'Spy': 'Action',
    'Road Trip': 'Adventure',
    'Jungle Adventure': 'Adventure',
    'Artificial Intelligence': 'Sci-Fi',
    'Body Horror': 'Horror',
    'Serial Killer': 'Crime',
    'True Crime': 'Crime',
    'Parody': 'Comedy',
    'Heist': 'Crime',
    'Monster Horror': 'Horror',
    'Swashbuckler': 'Adventure'
}
```

```python
def clean_main_genre(genre_list):
    if pd.isnull(genre_list):
        return 'Unknown'
    parts = genre_list.split(',')
    cleaned = []
    for part in parts:
        part = part.strip()
        if part in genre_mapping:
            cleaned.append(genre_mapping[part])
        else:
            cleaned.append(part)
    return ', '.join(cleaned)
```

```python
d1['Genre_Cleaned'] = d1['Main_Genre'].apply(clean_main_genre)
```

```python
print(d1['Genre_Cleaned'].head(20))
```

```python
def remove_duplicates(genre_string):
    if pd.isnull(genre_string):
        return 'Unknown'
    parts = genre_string.split(',')
    parts = [part.strip() for part in parts]
    parts = list(dict.fromkeys(parts))  # it Remove duplicates and keep it in order
    return ', '.join(parts)
```

```python
# Apply to Genre_Cleaned
d1['Genre_Cleaned'] = d1['Genre_Cleaned'].apply(remove_duplicates)
```

```python
#Code to Clean Gross Worldwide
d1['Gross Worldwide'] = d1['Gross Worldwide'].str.replace(',', '', regex=False)
```

```python
# Convert to numeric
d1['Gross Worldwide'] = pd.to_numeric(d1['Gross Worldwide'], errors='coerce')
```

```python
# # Filling missing values with 0 and convert to int
# d1['Gross Worldwide'] = d1['Gross Worldwide'].fillna(0).astype(int)
```

```python
#  Fill missing Gross Worldwide using Genre_Cleaned group median
d1['Gross Worldwide'] = d1['Gross Worldwide'].fillna(
    d1.groupby('Genre_Cleaned')['Gross Worldwide'].transform('median')
)

#  If still missing, fill using overall median
d1['Gross Worldwide'] = d1['Gross Worldwide'].fillna(
    d1['Gross Worldwide'].median()
)
```
